```python
import pandas as pd
import regex as re
import matplotlib.pyplot as plt
import nltk
import sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
from sklearn.manifold import TSNE
from gensim.models import Word2Vec
from pysentimiento import create_analyzer
from nltk.corpus import stopwords
```

```python
stocks = pd.read_csv("/content/stockerbot-export.csv")
nltk.download('stopwords')
stopwords = set(stopwords.words('english'))
stocks.head(5)
```

```python
def tokeniza(texto):
    tokens = re.findall(r'\b\w+\b', texto.lower())
    tokens_filtrados = [palabra for palabra in tokens if not re.search(r'[a-zA-Z]+\d+|\d+[a-zA-Z]+', palabra)]
    tokens_filtrados = [palabra for palabra in tokens_filtrados if not re.search(r'\W', palabra)]
    return tokens_filtrados

def quita_stopwords(palabras):
  return [palabra for palabra in palabras if palabra.lower() not in stopwords]

def crea_corpus(textos):
  corpus = []
  textos = textos.values.tolist()
  corpus = [palabra.lower() for oracion in textos for palabra in oracion]
  corpus = quita_stopwords(corpus)
  return corpus

def visualizar_palabra_tsne(modelo, palabra, tamanio):
    arreglo = np.empty((0, tamanio), dtype='f')
    etiquetas_palabras = [palabra]
    palabras_cercanas = modelo.wv.most_similar(palabra)
    arreglo = np.append(arreglo, np.array([modelo.wv[palabra]]), axis=0)

    for calificacion_palabra in palabras_cercanas:
        vector_palabras = modelo.wv[calificacion_palabra[0]]
        etiquetas_palabras.append(calificacion_palabra[0])
        arreglo = np.append(arreglo, np.array([vector_palabras]), axis=0)

    tsne = TSNE(n_components=2, perplexity=10, random_state=0)
    np.set_printoptions(suppress=True)
    Y = tsne.fit_transform(arreglo)
    x_coordenadas = Y[:, 0]
    y_coordenadas = Y[:, 1]

    plt.figure(figsize=(14, 8))
    plt.scatter(x_coordenadas, y_coordenadas)

    for etiqueta, x, y in zip(etiquetas_palabras, x_coordenadas, y_coordenadas):
        plt.annotate(etiqueta, xy=(x, y), xytext=(0, 0), textcoords='offset points')

    plt.xlim(x_coordenadas.min() - 0.0005, x_coordenadas.max() + 0.0005)
    plt.ylim(y_coordenadas.min() - 0.0005, y_coordenadas.max() + 0.0005)
    plt.show()

def visualizar_modeloo_TSNE(modelo,perplejidad, dimensiones, iteraciones):
    etiquetas = []
    tokens = []
    for palabra in modelo.wv.index_to_key:
          tokens.append(modelo.wv[palabra])
          etiquetas.append(palabra)
    tokens = np.array(tokens)
    tsne_modelo = TSNE(perplexity=perplejidad, init='pca', n_components=dimensiones, n_iter=iteraciones, verbose=1)
    nuevos_valores = tsne_modelo.fit_transform(tokens)
    x = []
    y = []
    for value in nuevos_valores:
        x.append(value[0])
        y.append(value[1])
    plt.figure(figsize=(32, 32))
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(etiquetas[i], xy=(x[i], y[i]), xytext=(5, 2), textcoords='offset points', ha='right', va='bottom')
    plt.show()
    va=('bottom')
    plt.show()
```

```python
mask = stocks["verified"] == "TRUE"
texto_verificado = stocks["text"][mask]
corpus = crea_corpus(texto_verificado.map(lambda oracion : tokeniza(oracion)))
corpus_verificado = crea_corpus(stocks["text"].map(lambda oracion : tokeniza(oracion)))
print(corpus)
```

```python
vectorizer = TfidfVectorizer(use_idf=True)
vectorizer_veri = TfidfVectorizer(use_idf=True)
vec = vectorizer.fit_transform(corpus)
vec_verificado = vectorizer_veri.fit_transform(corpus_verificado)
vec = pd.DataFrame(vec[0].T.todense(), index = vectorizer.get_feature_names_out(), columns=["tf-idf"])
vec_verificado = pd.DataFrame(vec_verificado[0].T.todense(), index = vectorizer_veri.get_feature_names_out(), columns=["tf-idf"])
vec = vec.sort_values('tf-idf', ascending=False)
vec_verificado_D = vec_verificado.sort_values('tf-idf', ascending=False)
print(vec.head(15))
print(vec_verificado_D.head(15))
```

```python
modelo = Word2Vec(sentences=[corpus], vector_size=100, window=5, min_count=1, workers=4)
visualizar_palabra_tsne(modelo,"money", 100)
visualizar_modeloo_TSNE(modelo, 10, 2, 500)
```

```python
def obtenProbabilidades(texto, sentimiento):
  negativos = []
  neutros = []
  positivos = []
  for probabilidad in sentimiento:
    negativos.append(probabilidad.probas['NEG'])
    neutros.append(probabilidad.probas['NEU'])
    positivos.append(probabilidad.probas['POS'])
  probabilidades = {'negativos': negativos, 'neutros': neutros, 'positivos' : positivos}
  probabilidades = pd.DataFrame.from_dict(probabilidades)
  return pd.concat([texto, probabilidades.reindex(texto.index)], axis = 1)
```

```python
analyzer = create_analyzer(task="sentiment", lang="en")
result = analyzer.predict(stocks["text"])
```

```python
Datos = obtenProbabilidades(stocks["text"], result)
```

```python
print("Negativo: ", Datos["negativos"].mean())
print("Neutro: ", Datos["neutros"].mean())
print("Positivo: ", Datos["positivos"].mean())
```

```python
mask = stocks["verified"] == "TRUE"
Datos_verificados = Datos[mask]
print("Negativo: ", Datos_verificados["negativos"].mean())
print("Neutro: ", Datos_verificados["neutros"].mean())
print("Positivo: ", Datos_verificados["positivos"].mean())
```

```python
Analisis = Datos_verificados["text"][Datos_verificados["positivos"] > 0.15]
Corpus_analisis = crea_corpus(Analisis.map(lambda oracion : tokeniza(oracion)))
```

```python
modelo_analisis = Word2Vec(sentences=[Corpus_analisis], vector_size=100, window=5, min_count=1, workers=4)
visualizar_palabra_tsne(modelo_analisis,"money", 100)
visualizar_modeloo_TSNE(modelo_analisis, 10, 2, 500)
```
